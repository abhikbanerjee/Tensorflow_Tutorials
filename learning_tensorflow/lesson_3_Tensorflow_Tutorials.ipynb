### Working with Images

```python
# reads an image at the location and prints the image shape
import matplotlib.image as mpimg
filename = "/Users/tkmacl9/Documents/"+"MarshOrchid.jpg"
image = mpimg.imread(filename)
print(image.shape)

# this can also be used to visualize the image using pyplot
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(image)
plt.show()
```

```python
# creating a transpose of the image using tensorflow
# image has the image read as a ndarray
import tensorflow as tf
x = tf.Variable(image , name ='x')
model = tf.global_variables_initializer()

with tf.Session() as session:
    # the perm method says to swap the 0th and 1st dimension and keep the second 
    # dimension unchanged
    x = tf.transpose(x, perm =[1,0,2])      
    session.run(model)
    result = session.run(x)

plt.imshow(result)
plt.show()
```

### Reverse variable length slices from the image

```python
import numpy as np
import os
# image is loaded from the previous cells
height, width, depth = image.shape

# Create a TensorFlow Variable
x = tf.Variable(image, name='x')

model = tf.global_variables_initializer()

with tf.Session() as session:
    # tf.reverse_sequence(input, seq_lengths, seq_dim, batch_dim=None, name=None)
    # starting with the batch dimension take seuence lengths from the sequence dimension and reverse them
    x = tf.reverse_sequence(x, [width] * height, 1, batch_dim=0)
    session.run(model)
    result = session.run(x)

print(result.shape)
plt.imshow(result)
plt.show()
```
